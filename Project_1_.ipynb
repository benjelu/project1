```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,10])
y = np.array([2,4,10])
plt.plot(x, y, color = 'red',  linewidth = 3, label = 'did not study')             
plt.plot(x, y**2, color = 'blue', linewidth = 3, label = 'study')
plt.title("Percentage of knowledge with or without study",fontdict={'fontname':'Arial Black','fontsize':16})
plt.legend(loc='upper right')
plt.xlabel("Time")
plt.ylabel("Percentage")
plt.plot(x, y)
plt.savefig("pycs4001-05.png")
```
